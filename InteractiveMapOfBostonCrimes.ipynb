## Part 1: Altair interactive plots

Gapminder is a non-profit organization that promotes global sustainable development and seeks to bridge the gap between misconceptions and data-driven understanding. We are going to explore the a small subset of its data with the information of average income, health score and population for each country in the world. The region information is also provided. 

```python
import altair as alt
import pandas as pd
import geopandas as gpd

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)


import zipfile

with zipfile.ZipFile("crime.csv.zip") as z:
    # List all files inside the ZIP
    for name in z.namelist():
        print(name)

    # Pick the actual CSV (not the __MACOSX one)
    with z.open("crime - crime.csv.csv") as f:
        df = pd.read_csv(f)

# Now df is your DataFrame
df.head()
```

```python
from shapely.geometry import Point

#Convert to geodataframe

df = df.dropna(subset=["Lat", "Long"])  # remove rows without coordinates
geometry = [Point(xy) for xy in zip(df["Long"], df["Lat"])]
gdf = gpd.GeoDataFrame(df, geometry=geometry, crs="EPSG:4326")

gdf.head()
```

```python
districts = gpd.read_file("boston_police_districts.geojson").to_crs(epsg=4326)

# map each district to it's name

district_mapping = {
    'A1': 'Charlestown',
    'A15': 'Downtown',
    'A7': 'East Boston',
    'B2': 'Roxbury',
    'B3': 'Mattapan',
    'C11': 'Dorchester',
    'C6': 'South Boston',
    'D14': 'Brighton',
    'D4': 'South End',
    'E13': 'Jamaica Plain',
    'E18': 'Hyde Park',
    'E5': 'West Roxbury'
}


districts['DISTRICT_NAME'] = districts['DISTRICT'].map(district_mapping)
districts.head()
```

```python
joined = gpd.sjoin(gdf, districts, how="left", predicate="within")

joined.head()
```

```python
# AGGREGATE

counts = joined.groupby("DISTRICT_").size().reset_index(name="crime_count")

counts.head()
```

```python
import folium

# Merge back with the district shapes
districts = districts.merge(counts, on="DISTRICT_", how="left")
districts["crime_count"] = districts["crime_count"].fillna(0)
```

```python
# Plot choropleth map

districts.rename(columns={
            "DISTRICT_NAME": "District",
            "crime_count": "Crime Count"
        })

m = districts.explore(
    tiles="CartoDB positron",
    column="Crime Count",
    cmap="OrRd",
    legend=True,
    tooltip=["District", "Crime Count"],
    name="Crime Counts",
    height = "800px"
)

folium.LayerControl().add_to(m)
m
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import folium

offense_options = sorted(joined["OFFENSE_CODE_GROUP"].dropna().unique())

offense_dropdown = widgets.Dropdown(
    options=offense_options,
    value=offense_options[0],
    description='Offense:',
    layout=widgets.Layout(width='60%')
)

out = widgets.Output()

def update_map(change):
    with out:
        clear_output(wait=True)
        
        selected_offense = change['new']
        
        # Filter and aggregate
        filtered = joined[joined["OFFENSE_CODE_GROUP"] == selected_offense]
        counts = filtered.groupby("DISTRICT_").size().reset_index(name="crime_count")

        # Clean + merge counts with district polygons
        updated = districts.drop(columns="crime_count", errors="ignore")
        updated = updated.merge(counts, on="DISTRICT_", how="left")
        updated["crime_count"] = updated["crime_count"].fillna(0)
        updated = updated.rename(columns={
            "DISTRICT_NAME": "District",
            "crime_count": "Crime Count"
        })

        # Plot map
        m = updated.explore(
            column="Crime Count",
            cmap="OrRd",
            legend=True,
            tooltip=["District", "Crime Count"],
            name=f"{selected_offense} Crimes",
            tiles="CartoDB positron",
            height="700px"
        )
        folium.LayerControl().add_to(m)
        display(m)

offense_dropdown.observe(update_map, names='value')

display(offense_dropdown, out)
update_map({'new': offense_dropdown.value})  # show initial map
```
