```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import altair as alt

alt.data_transformers.enable('default', max_rows=None)

import zipfile

zip_path = 'crime.csv.zip'  # Replace with your actual ZIP file path

with zipfile.ZipFile(zip_path, 'r') as z:
    csv_filename = z.namelist()[0]
    
    with z.open(csv_filename) as f:
        df = pd.read_csv(f)
df 
```

```python
grouped = df.groupby(['YEAR', 'MONTH', 'OFFENSE_CODE_GROUP']).size().reset_index(name='COUNT')
grouped['DATE'] = pd.to_datetime(grouped[['YEAR', 'MONTH']].assign(DAY=1)).dt.strftime('%Y-%m')


grouped.to_csv('grouped_crime_data.csv', index=False)
```
